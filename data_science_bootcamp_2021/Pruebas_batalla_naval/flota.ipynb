```python
 # Pedir la ubicacion al usuario
   
    def place_ship(board, ship, orientacion, fila, columna):
        
        # orientacion de los barcos
        if orientacion == "v":
            for i in range(ship):
                board[fila + i][columna]
        elif orientacion == "h":
            for i in range(ship):
                board[fila][columna + i]

        return board
```
