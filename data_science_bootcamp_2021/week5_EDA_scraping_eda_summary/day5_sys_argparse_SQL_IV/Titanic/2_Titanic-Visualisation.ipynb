6. Using matplotlib find the appropriate plot to visualize the column `'gender'` 

```python
# your code here
fig, ax = plt.subplots()  
x = titanic.gender.unique()  
y = titanic.gender.value_counts()  
ax.bar(x, y)  
ax.set_xlabel('Gender')  
ax.set_ylabel('Count')  
plt.show() 
```
